```python
import numpy as np
import pandas as pd
from ast import literal_eval
```

```python
path = "./Data/search_grid.csv"
df = pd.read_csv(path, converters={"mapped_zipcodes": literal_eval})
```

```python
df = df.explode("mapped_zipcodes").reset_index()
df['mapped_zipcodes'] = pd.to_numeric(df['mapped_zipcodes'])
```
